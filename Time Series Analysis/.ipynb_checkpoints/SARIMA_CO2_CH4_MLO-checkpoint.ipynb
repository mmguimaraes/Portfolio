# SARIMA model applied to CO<sub>2</sub> and CH<sub>4</sub> data collected at Mauna Loa Observatory. 

SARIMA stands for Seasonal Autoregressive Integrated Moving Averages and it is a very useful technique applied to time series analysis, specially with the data presents seasonality.

We will take a look at the data collected at the Mauna Loa Observatory, in Hawaii, for two important Greenhouse gases: CO<sub>2</sub> and CH<sub>4</sub>. 

The data was retrieved from the Earth System Research Laboratory (ESRL) Global Monitoring Division website: <a href='https://www.esrl.noaa.gov/gmd/dv/data/index.php'>ESRL/GMD Data Finder.</a> 

If you want to know more about SARIMA, take a look at:
<a href='https://www.statsmodels.org/stable/statespace.html'>Statsmodels Tutorial</a>

### Perform standard imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

# Load specific forecasting tools
# Although we are using a SARIMA model, the statsmodel implementation is called SARIMAX. 
# The X stands for eXogenous data.
from statsmodels.tsa.statespace.sarimax import SARIMAX

from statsmodels.tsa.seasonal import seasonal_decompose      # for ETS Plots
from pmdarima import auto_arima                              # for determining ARIMA orders

# Ignore harmless warnings
import warnings
warnings.filterwarnings("ignore")
```


### Load datasets and perform some adjustments

```python
# CO2 dataframe
co2df = pd.read_csv('data/co2_mlo_monthly_data.csv',sep=' ')
co2df.head()
```

```python
# CH4 dataframe
ch4df = pd.read_csv('data/ch4_mlo_monthly_data.csv',sep=' ')
ch4df.head()
```

### There are some columns without use at the moment: site_code, latitude, longitude, altitute, elevation, intake_height, qcflag.
### Because we are using a Monthly averaged dataset, we don't need the columns for hour, minute and second.
### We will drop all theses columns to clean the dataframe.

```python
list_drop = 'site_code hour minute second latitude longitude altitude elevation intake_height qcflag'.split()
co2df.drop(list_drop,axis=1,inplace=True)
ch4df.drop(list_drop,axis=1,inplace=True)
co2df.head()
```

```python
ch4df.head()
```

### Now we need to combine the columns year, month and day into a DatetimeIndex.
### We can do this by passing a dictionary into <tt>pandas.to_datetime()</tt> with year, month and day values.

```python
# Add a "date" datetime column
co2df['date']=pd.to_datetime(dict(year=co2df['year'], month=co2df['month'], day=co2df['day']))
ch4df['date']=pd.to_datetime(dict(year=ch4df['year'], month=ch4df['month'], day=ch4df['day']))
co2df.head()
```

```python
ch4df.head()
```

### We will set the date column as the index of the dataframe

```python
co2df.set_index('date',inplace=True)
ch4df.set_index('date',inplace=True)
co2df.head()
```

```python
ch4df.head()
```

### Let's take a look a the index

```python
co2df.index
```

```python
ch4df.index
```

### The frequency is set to None. We have to set it to MS (Month Start). 
### We can do it using the <tt>.index.freq='MS'</tt> method. 

```python
ch4df.index.freq = 'MS'
ch4df.index
```

### But if we try this with the CO<sub>2</sub> dataframe, we will receive an error: <tt>Inferred frequency None from passed values does not conform to passed frequency MS</tt>.
### This is happening because there is 1 date mising from the dateset, specifically the date 1975-11-1. 
### We need the datetime to be equally spaced.
### We will use the method <tt> asfreq()</tt> from Pandas to fill this date.

```python
co2df = co2df.asfreq(freq='MS',method='ffill')
```

```python
co2df.index.freq = 'MS'
co2df.index
```

```python
title = 'Monthly Mean CO‚ÇÇ Levels (ppm) over Mauna Loa, Hawaii'
ylabel='parts per million'
xlabel='Date'

ax = co2df['value'].plot(figsize=(12,6),title=title)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

```python
title = 'Monthly Mean CH$_4$ Levels (ppb) over Mauna Loa, Hawaii'
ylabel='parts per billion'
xlabel='Date'

ax = ch4df['value'].plot(figsize=(12,6),title=title)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

### The next step is the determine the orders (p,d,q) for the SARIMA model.
### We will use <tt>auto_arima</tt> for this job.

## Run <tt>pmdarima.auto_arima</tt>
### For SARIMA Orders we set seasonal=True and pass in an m value equal to 12.

### First we do it for the CO<sub>2</sub> data.

```python
auto_arima(co2df['value'],seasonal=True,m=12).summary()
```

### The result is an ARIMA Order of (1,1,2) combined with a seasonal order of (2,0,2,12).

### Now let's do it for the CH4 data.

```python
auto_arima(ch4df['value'],seasonal=True,m=12).summary()
```

### The result is an ARIMA Order of (1,1,3) combined with a seasonal order of (1,0,1,12).

### Now let's train and test these models.

### Split the data into train/test sets

```python
# We will define a variable called nobs (number of observations) and set it to one year for testing.
# If we want to change it later, the code will keep working.
nobs=12
# Define the train set
co2_train = co2df.iloc[:-nobs]
ch4_train = ch4df.iloc[:-nobs]
# Define the test set 
co2_test = co2df.iloc[-nobs:]
ch4_test = ch4df.iloc[-nobs:]
# Check the length of the train and test sets.
print('The length of the CO2 train set is: ',len(co2_train))
print('The length of the CO2 test set is: ',len(co2_test))
print('\n')
print('The length of the CH4 train set is: ',len(ch4_train))
print('The length of the CH4 test set is: ',len(ch4_test))
```

### Now we will call the <tt>SARIMAX</tt> model and fit it to the train dataset.

```python
co2_model = SARIMAX(co2_train['value'],order=(1,1,2),seasonal_order=(2,0,2,12))
co2_results = co2_model.fit()
co2_results.summary()
```

```python
ch4_model = SARIMAX(ch4_train['value'],order=(1,1,3),seasonal_order=(1,0,1,12))
ch4_results = ch4_model.fit()
ch4_results.summary()
```

### Now we will obtain the predicted values

```python
# Define the start point
co2_start=len(co2_train)
ch4_start=len(ch4_train)
# Define the end point
co2_end=len(co2_train)+len(co2_test)-1
ch4_end=len(ch4_train)+len(ch4_test)-1
co2_predictions = co2_results.predict(start=co2_start, end=co2_end, dynamic=False, typ='levels').rename('SARIMA(1,1,2)(2,0,2,12) Predictions')
ch4_predictions = ch4_results.predict(start=ch4_start, end=ch4_end, dynamic=False, typ='levels').rename('SARIMA(1,1,3)(1,0,1,12) Predictions')
```

Passing <tt>dynamic=False</tt> means that forecasts at each point are generated using the full history up to that point (all lagged values).

Passing <tt>typ='levels'</tt> predicts the levels of the original endogenous variables. If we'd used the default <tt>typ='linear'</tt> we would have seen linear predictions in terms of the differenced endogenous variables.

For more information on these arguments visit https://www.statsmodels.org/stable/generated/statsmodels.tsa.arima_model.ARIMAResults.predict.html

### Because now we have the predicted and the real values, we can compare them!

```python
# Compare predictions to expected values
print('CO2 Predictions and Expected values')
for i in range(len(co2_predictions)):
    print(f"predicted={co2_predictions[i]:<11.8}, expected={co2_test['value'][i]}, difference={co2_test['value'][i]-co2_predictions[i]:<10.5}")
```

```python
# Compare predictions to expected values
print('CH4 Predictions and Expected values')
for i in range(len(ch4_predictions)):
    print(f"predicted={ch4_predictions[i]:<11.8}, expected={ch4_test['value'][i]}, difference={ch4_test['value'][i]-ch4_predictions[i]:<10.5}")
```

### Let's take a look at the plot for this section of the dataset.

```python
title = 'Monthly Mean CO‚ÇÇ Levels (ppm) over Mauna Loa, Hawaii'
ylabel='ppm'
xlabel=''

ax = co2_test['value'].plot(legend=True,figsize=(12,8),title=title)
co2_predictions.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

```python
title = 'Monthly Mean CH$_4$ Levels (ppb) over Mauna Loa, Hawaii'
ylabel='ppb'
xlabel=''

ax = ch4_test['value'].plot(legend=True,figsize=(12,8),title=title)
ch4_predictions.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

### Our model seems to behave very well, predicting the seasonality of the data.

### Now let's evaluate the model using the Mean Squared Error and the Root Mean Squared Error.

```python
from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse

ms_error = mean_squared_error(co2_test['value'], co2_predictions)
rms_error = rmse(co2_test['value'], co2_predictions)

print('CO2')
print('\n')
print(f'SARIMA(1,1,2)(2,0,2,12) RMSE Error: {ms_error:11.10}')
print(f'SARIMA(1,1,2)(2,0,2,12) MSE Error: {rms_error:11.10}')
```

```python
ms_error = mean_squared_error(ch4_test['value'], ch4_predictions)
rms_error = rmse(ch4_test['value'], ch4_predictions)

print('CH4')
print('\n')
print(f'SARIMA(1,1,3)(1,0,1,12) RMSE Error: {ms_error:11.10}')
print(f'SARIMA(1,1,3)(1,0,1,12) MSE Error: {rms_error:11.10}')
```

### Now for the forecasting part!

### We will retrain our models using the full data, and forecast the near future (12 months).

```python
#CO2
co2_model = SARIMAX(co2df['value'],order=(1,1,2),seasonal_order=(2,0,2,12))
co2_results = co2_model.fit()
co2_fcast = co2_results.predict(len(co2df),len(co2df)+12,typ='levels').rename('SARIMA(1,1,2)(2,0,2,12) Forecast')

#CH4
ch4_model = SARIMAX(ch4df['value'],order=(1,1,3),seasonal_order=(1,0,1,12))
ch4_results = ch4_model.fit()
ch4_fcast = ch4_results.predict(len(ch4df),len(ch4df)+12,typ='levels').rename('SARIMA(1,1,3)(1,0,1,12) Forecast')
```

```python
title = 'Monthly Mean CO‚ÇÇ Levels (ppm) over Mauna Loa, Hawaii'
ylabel='ppm'
xlabel=''

ax = co2df['value'].plot(legend=True,figsize=(12,6),title=title)
co2_fcast.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

```python
title = 'Monthly Mean CH$_4$ Levels (ppb) over Mauna Loa, Hawaii'
ylabel='ppb'
xlabel=''

ax = ch4df['value'].plot(legend=True,figsize=(12,6),title=title)
ch4_fcast.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```
