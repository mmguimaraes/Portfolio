# SARIMA(p,d,q)(P,D,Q)m
# Seasonal Autoregressive Integrated Moving Averages
We have finally reached one of the most fascinating aspects of time series analysis: seasonality.

Where ARIMA accepts the parameters $(p,d,q)$, SARIMA accepts an <em>additional</em> set of parameters $(P,D,Q)m$ that specifically describe the seasonal components of the model. Here $P$, $D$ and $Q$ represent the seasonal regression, differencing and moving average coefficients, and $m$ represents the number of data points (rows) in each seasonal cycle.

<strong>NOTE:</strong> The statsmodels implementation of SARIMA is called SARIMAX. The “X” added to the name means that the function also supports <em>exogenous</em> regressor variables. We'll cover these in the next section.

## Perform standard imports and load datasets

```python
import pandas as pd
import numpy as np
%matplotlib inline

# Load specific forecasting tools
from statsmodels.tsa.statespace.sarimax import SARIMAX

from statsmodels.graphics.tsaplots import plot_acf,plot_pacf # for determining (p,q) orders
from statsmodels.tsa.seasonal import seasonal_decompose      # for ETS Plots
from pmdarima import auto_arima                              # for determining ARIMA orders

# Ignore harmless warnings
import warnings
warnings.filterwarnings("ignore")

# Load dataset
df = pd.read_csv('mlo_N2O_MM.csv')
```

### Inspect the data, create a DatetimeIndex

```python
df.head()
```

We need to combine two integer columns (year and month) into a DatetimeIndex. We can do this by passing a dictionary into <tt>pandas.to_datetime()</tt> with year, month and day values.<br>
For more information visit https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html

```python
# Add a "date" datetime column
df['date']=pd.to_datetime(dict(year=df['year'], month=df['mon'], day=1))
```

```python
df.head()
```

```python
df.index
```

```python
title = 'Monthly Mean N2O Levels (ppm) over Mauna Loa, Hawaii'
ylabel='parts per billion'
xlabel='' # we don't really need a label here

ax = df['concentration in ppb'].plot(figsize=(12,6),title=title,marker='*')
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

## We cannot set the Freq to MS because there missing values. We don't have the data for the period from 1991-09-01 to 1992-01-01.

## We will separate the dataset into 2 different datasets. One containing data until 1991-08-01 and one containing data from 1992-02-01 until the end.

## We will then create a SARIMA model to predict the values we are missing and later join the 2 datasets.

```python
df1 = df[df.date <= '1991-08-01']
df2 = df[df.date > '1991-08-01']
```

```python
# Set "date" to be the index
df1.index.freq = 'MS'
df1.index
```

```python
df2.index.freq= 'MS'
df2.index
```

### Plot the source data

```python
title = 'Monthly Mean N2O Levels (ppm) over Mauna Loa, Hawaii'
ylabel='parts per billion'
xlabel='' # we don't really need a label here

ax = df1['concentration in ppb'].plot(figsize=(12,6),title=title)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

### Run an ETS Decomposition

```python
result = seasonal_decompose(df['value'], model='add')
result.plot();
```

Although small in scale compared to the overall values, there is a definite annual seasonality.

### Run <tt>pmdarima.auto_arima</tt> to obtain recommended orders
This may take awhile as there are a lot more combinations to evaluate.

```python
# For SARIMA Orders we set seasonal=True and pass in an m value
auto_arima(df['value'],seasonal=True,m=12).summary()
```

Excellent! This provides an ARIMA Order of (0,1,3) combined with a seasonal order of (1,0,1,12) Now let's train & test the SARIMA(0,1,3)(1,0,1,12) model, evaluate it, then produce a forecast of future values.
### Split the data into train/test sets

```python
len(df)
```

```python
# Set one year for testing
nobs=12
train = df.iloc[:-nobs]
test = df.iloc[-nobs:]
len(test)
```

### Fit a SARIMA(0,1,3)(1,0,1,12) Model

```python
model = SARIMAX(train['value'],order=(1,1,3),seasonal_order=(1,0,1,12))
results = model.fit()
results.summary()
```

```python
# Obtain predicted values
start=len(train)
end=len(train)+len(test)-1
predictions = results.predict(start=start, end=end, dynamic=False, typ='levels').rename('SARIMA(1,1,3)(1,0,1,12) Predictions')
```

Passing <tt>dynamic=False</tt> means that forecasts at each point are generated using the full history up to that point (all lagged values).

Passing <tt>typ='levels'</tt> predicts the levels of the original endogenous variables. If we'd used the default <tt>typ='linear'</tt> we would have seen linear predictions in terms of the differenced endogenous variables.

For more information on these arguments visit https://www.statsmodels.org/stable/generated/statsmodels.tsa.arima_model.ARIMAResults.predict.html

```python
# Compare predictions to expected values
for i in range(len(predictions)):
    print(f"predicted={predictions[i]:<11.10}, expected={test['value'][i]}")
```

```python
# Plot predictions against known values
title = 'Monthly Mean CH4 Levels (ppm) over Mauna Loa, Hawaii'
ylabel='parts per million'
xlabel=''

ax = test['value'].plot(legend=True,figsize=(12,6),title=title)
predictions.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

### Evaluate the Model

```python
from sklearn.metrics import mean_squared_error

error = mean_squared_error(test['value'], predictions)
print(f'SARIMA(1,1,3)(1,0,1,12) MSE Error: {error:11.10}')
```

```python
from statsmodels.tools.eval_measures import rmse

error = rmse(test['value'], predictions)
print(f'SARIMA(1,1,3)(1,0,1,12) RMSE Error: {error:11.10}')
```

These are outstanding results!
### Retrain the model on the full data, and forecast the future

```python
model = SARIMAX(df['value'],order=(1,1,3),seasonal_order=(1,0,1,12))
results = model.fit()
fcast = results.predict(len(df),len(df)+11,typ='levels').rename('SARIMA(1,1,3)(1,0,1,12) Forecast')
```

```python
# Plot predictions against known values
title = 'Monthly Mean CH4 Levels (ppm) over Mauna Loa, Hawaii'
ylabel='parts per million'
xlabel=''

ax = df['value'].plot(legend=True,figsize=(12,6),title=title)
fcast.plot(legend=True)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```
